```python
# Import necessary modules
import numpy as np
import pandas as pd

# Declare the length required for the parameters list
DEFAULT_LENGTH = 11

# Loop repeatedly to ask the user if they want to choose a custom or default dataset
while True:
    # Prompt the user to choose between a default dataset or inputting a custom one
    inputuser = input("Hello, would you want to enter a custom training dataset or use a default dataset (type 'custom' or 'default'): ")
    
    # Check if the user enterred 'custom'
    if inputuser.lower() == "custom":
        # Create an empty list to store the arrays
        arrays = []

        # Loop repeatedly to show the user the available parameters and the chance to input parameters
        while True:
            
            # Display the available parameters to the user and the options to choose from
            print("""
                parameter 1: Is there an alternate restaurant? ('yes' or 'no')
                parameter 2: Is there a bar in the restaurant? ('yes' or 'no')
                parameter 3: Is it Friday? ('yes' or 'no')
                parameter 4: Are you hungry? ('yes' or 'no')
                parameter 5: Are there patrons? ('none' or 'some' or 'full')
                parameter 6: Is it expensive? ('$' for cheap, '$$' for affordable, '$$$' for expensive) 
                parameter 7: Is it raining? ('yes' or 'no')
                parameter 8: Do you have a reservation? ('yes' or 'no')
                parameter 9: Which type of food are they serving? ('french', 'italian', 'thai' or 'burger')
                parameter 10: What is the estimated waiting time? ('0-10', '10-30', '30-60', '>60')
                parameter 11: Will you wait? (yes or no)
            """)
            # Prompt the user to enter the parameters separated by spaces or stop
            user_input = input("Enter several parameters separated by spaces (or type 'stop' to stop): ").lower()

            # Check if the user wants to stop and the array list is valid
            if user_input.lower() == "stop" and len(arrays)>3:
                break
            # Check if the user wants to stop and the array list is not valid     
            elif user_input.lower() == "stop" and len(arrays)<=3:
                print("We need at least 4 entries to be able to make prediction. Please enter more data.")
                continue

            # Split the input string into a list of parameters
            parameters = user_input.split()
            # Validate the parameters
            valid = True
            
            # Validate the length of the parameter
            if len(parameters)<DEFAULT_LENGTH or len(parameters)>DEFAULT_LENGTH:
                print(f"\n-*-*-*-*-*-*-The data set is not in the right format, you must enter {DEFAULT_LENGTH} parameters-*-*-*-*-*-*-\n")
                continue
            
            # Validate values of every parameter that could be inputted
            valid_values = [
                ["yes", "no"],
                ["yes", "no"],
                ["yes", "no"],
                ["yes", "no"],
                ["none", "some", "full"],
                ["$", "$$", "$$$"],
                ["yes", "no"],
                ["yes", "no"],
                ["french", "italian", "thai", "burger"],
                ["0-10", "10-30", "30-60", ">60"],
                ["yes", "no"]
            ]
            
            # Validate the inputs
            for i, param in enumerate(parameters):
                if param not in valid_values[i]:
                    valid = False
                    break
                    
            # Print an 'Invalid Input' error if the the input was not valid        
            if not valid:    
                print("\n-*-*-*-*-*-*-*-Invalid input. Please enter the parameters again-*-*-*-*--*-*-*-*-*.\n")
                continue
            
            # Create a NumPy array from the parameters list
            arr = np.array(parameters)

            # Append the array to the 'Arrays' list
            arrays.append(arr)
            

            
                   
                
        # Create a NumPy array from the list of arrays
        dataset = np.array(arrays)
        break
     
    # Check if the user enterred 'default', so the following default dataset will be assigned
    elif inputuser.lower() == "default":
        dataset = np.array([
['yes', 'no', 'no', 'yes', 'some', '$$$', 'no', 'yes', 'french', '0-10', 'yes'],
['yes', 'no', 'no', 'yes', 'full', '$', 'no', 'no', 'thai', '30-60', 'no'],
['no', 'yes', 'no', 'no', 'some', '$', 'no', 'no', 'burger', '0-10', 'yes'],
['yes', 'no', 'yes', 'yes', 'full', '$', 'yes', 'no', 'thai', '10-30', 'yes'],
['yes', 'no', 'yes', 'no', 'full', '$$$', 'no', 'yes', 'french', '>60', 'no'],
['no', 'yes', 'no', 'yes', 'some', '$$', 'yes', 'yes', 'italian', '0-10', 'yes'],
['no', 'yes', 'no', 'no', 'none', '$', 'yes', 'no', 'burger', '0-10', 'no'],
['no', 'no', 'no', 'yes', 'some', '$$', 'yes', 'yes', 'thai', '0-10', 'yes'],
['no', 'yes', 'yes', 'no', 'full', '$', 'yes', 'no', 'burger', '>60', 'no'],
['yes', 'yes', 'yes', 'yes', 'full', '$$$', 'no', 'yes', 'italian', '10-30', 'no'],
['no', 'no', 'no', 'no', 'none', '$', 'no', 'no', 'thai', '0-10', 'no'],
['yes', 'yes', 'yes', 'yes', 'full', '$', 'no', 'no', 'burger', '30-60', 'yes'],
])
        
        break
    # Print an error message in case of an input error
    else:
        print("Please enter 'custom' or 'default'")
# Create a pandas dataframe with the dataset declared and print it for the user
df2 = pd.DataFrame(dataset, columns=['Alt', 'Bar', 'Fri','Hun', 'Pat', 'Price','Rain', 'Res', 'Type','Est', 'WillWait'])
blankIndex=[''] * len(df2)
df2.index=blankIndex

print(df2)
```

```python
# Import necessary modules
from sklearn import tree
from sklearn import preprocessing

# Extract inputs and output from the dataset
x = dataset[:,0:10]
y = dataset[:, 10]

# Encode inputs using LabelEncoder
le = preprocessing.LabelEncoder()
for i in range(0,10):
    x[:,i] = le.fit_transform(x[:,i])
    
# Encode the output using LabelEncoder 
y = le.fit_transform(y)
```

```python
# Create a decision tree classifier object
dtc = tree.DecisionTreeClassifier(criterion="entropy")
```

```python
# Build the decision tree classifier using inputs (x) and output (y)
dtc.fit(x,y)
```

```python
# Plot the decision tree
tree.plot_tree(dtc)
```

```python
# Declare the length required for the parameters list
DEFAULT_LENGTH_PRED =10

another_predict = ""

# Loop repeatedly to ask the user if they want to make a prediction
while True:
    
    will_predict = input(f"Do you want to make a {another_predict}prediction based on the current dataset ? (yes or no)")
    
    # Print 'another' inside the previous input statement after the first loop happen
    another_predict = "another "
    
    # Validate input
    if will_predict.lower() not in ("yes","no"):
        print("Invalid input. Please enter yes or no")
        continue
    
    # Check if user inputs 'no'
    if will_predict.lower() == "no":
        break

    
    # Print the available parameters and the values corresponding to each of them
    print("""
            parameter 1: Is there an alternate restaurant? ( 1 for 'yes' or 0 for 'no')
            parameter 2: Is there a bar in the restaurant? (1 for 'yes' or 0 for 'no')
            parameter 3: Is it Friday? (1 for 'yes' or 0 for 'no')
            parameter 4: Are you hungry? (1 for 'yes' or 0 for 'no')
            parameter 5: Are there patrons? (1 for 'none' or 2 for'some' or 0 for 'full')
            parameter 6: Is it expensive? (0 for '$', 1 for '$$', 2 for '$$$') 
            parameter 7: Is it raining? (1 for 'yes' or 0 for 'no')
            parameter 8: Do you have a reservation? (1 for 'yes' or 0 for 'no')
            parameter 9: Which type of food are they serving? (1 for 'french', 2 for 'italian', 3 for 'thai' or 0 for 'burger')
            parameter 10: What is the estimated waiting time? (0 for '0-10', 1 for '10-30', 2 for '30-60', 3 for '>60')
        """)

    # Prompt the user to enter parameter values
    user_input = input("Enter several parameters separated by spaces (or type 'stop' to stop): ").lower()
    

    # Split the input string into a list of parameters
    parameters = user_input.split()
    
    # Validate the parameters
    valid = True

    # Validate the length of the list
    if len(parameters)<DEFAULT_LENGTH_PRED or len(parameters)>DEFAULT_LENGTH_PRED:
        print(f"\n-*-*-*-*-*-*-The data set is not in the right format, you must enter {DEFAULT_LENGTH_PRED} parameters-*-*-*-*-*-*-\n")
        continue
        
        # Validate the values
    valid_values = [
        ["1", "0"],
        ["1", "0"],
        ["1", "0"],
        ["1", "0"],
        ["1", "2", "0"],
        ["0", "1", "2"],
        ["1", "0"],
        ["1", "0"],
        ["0", "1", "2", "3"],
        ["0", "1", "2", "3"],
    ]
    
    for i, param in enumerate(parameters):
        if param not in valid_values[i]:
            valid = False
            break

    # Print error if input is not valid
    if not valid:    
        print("\n-*-*-*-*-*-*-*-Invalid input. Please enter the parameters again-*-*-*-*--*-*-*-*-*.\n")
        continue
    
    # Return the predicted label of what the output will be in reference to the constructed tree
    y_pred = dtc.predict([parameters])
    
    # Inverse transform back the predicted label to the original wording representation and print it
    print("Predicted output: ", le.inverse_transform(y_pred))
```

```python
# Prompt the user to choose to get an improved decision tree and classification report or not
inp = input("Would you like to receive an improved and clearer version of the decision tree (in PDF format), \
along with a classification report that demonstrates the accuracy of the tree? (type '1' if interested): ")

# Check if the user wants the improved decision tree
if inp == '1':
    # Import the necessary modules
    import graphviz
    from sklearn.metrics import classification_report

    # Generate the graphviz representation of the decision tree
    dot_data = tree.export_graphviz(dtc, out_file=None,
                                    feature_names=['Alt', 'Bar', 'Fri', 'Hun', 'Pat', 'Price', 'Rain', 'Res', 'Type', 'Est'],
                                    class_names=le.classes_,
                                    filled=True, rounded=True)
    graph = graphviz.Source(dot_data)
    
    # Render the decision tree as a PDF file
    graph.render("Decision_tree_default")

    # Predict the target variable based on the decision tree
    y_p = dtc.predict(x)

    # Generate a classification report showing the accuracy of the decision tree
    print(classification_report(y, y_p))
    
    # Print a thank you message
    print("\n***---***---***---***Thank you for using our program!.***---***---***---***")
    
else:
    # Print a thank you message if the user does not want the decision tree and report
    print("\n***---***---***---***Thank you for using our program!.***---***---***---***")
```
